```python
import zipfile

with zipfile.ZipFile('Sample - Superstore.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()

print("File Extracted Successfully!")
```

```python
import dask.dataframe as dd

df = dd.read_csv(
    "Sample - Superstore.csv",
    encoding="latin1"
)

print(df.head())
```

```python
# Dataset information
print("Number of rows:", df.shape[0].compute())
print("Number of columns:", len(df.columns))
```

```python
# Total sales by category
sales_by_category = df.groupby("Category")["Sales"].sum().compute()
print(sales_by_category)
```

```python
# Total profit by region
profit_by_region = df.groupby("Region")["Profit"].sum().compute()
print(profit_by_region)
```

```python
# Top 10 products by sales
top_products = df.groupby("Product Name")["Sales"].sum().nlargest(10).compute()
print(top_products)
```
